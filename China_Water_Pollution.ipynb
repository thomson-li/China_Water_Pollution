## Weighted calculate new wqi, compare with the wqi dataset given

```python
import pandas as pd

# read dataset
df = pd.read_csv(r'/Users/lihaocong/Documents/Introtuction to Data Science/china_water_pollution_data.csv')

# define weight
weights = {
    'pH': 0.1,
    'Dissolved_Oxygen_mg_L': 0.2,
    'COD_mg_L': 0.15,
    'BOD_mg_L': 0.15,
    'Total_Phosphorus_mg_L': 0.1,
    'Total_Nitrogen_mg_L': 0.1,
    'Turbidity_NTU': 0.05,
    'Nitrate_mg_L': 0.05,
    'Nitrite_mg_L': 0.05,
    'Ammonia_N_mg_L': 0.05
}

# Keep only the columns you need
cols = list(weights.keys())
df_weighted = df[cols].copy()

# Normalization (0-1 standardization to prevent dimension effects)
for col in cols:
    min_val = df_weighted[col].min()
    max_val = df_weighted[col].max()
    if max_val > min_val:
        df_weighted[col] = (df_weighted[col] - min_val) / (max_val - min_val)
    else:
        df_weighted[col] = 0  
    # If the maximum value is equal to the minimum value, all values ​​are assigned to 0 to avoid division by zero errors

# Calculate the weighted score
df_weighted['Comprehensive water quality score'] = sum(df_weighted[col] * w for col, w in weights.items())

# generate new csv
df_weighted.to_csv('weighted_water_quality_score.csv', index=False)
print(df_weighted.head())
```

```python
# Line chart: WQI change curve of Beijing and Chengdu over time (one month)
import pandas as pd
import matplotlib.pyplot as plt

# Read the files of two cities (after data processing, only one piece of data is retained for multiple measurements on the same day)
df_bj = pd.read_csv(r'/Users/lihaocong/Documents/Introtuction to Data Science/Beijing_water_pollution_data.csv')
df_cd = pd.read_csv(r'/Users/lihaocong/Documents/Introtuction to Data Science/Chengdu_water_pollution_data.csv')

df_bj['City'] = 'beijing'
df_cd['City'] = 'chengdu'

df = pd.concat([df_bj, df_cd], ignore_index=True)

# Standardize column formats
df['City'] = df['City'].str.lower().str.strip()
df['Date'] = pd.to_datetime(df['Date'])

# Get data for the last 30 days
latest_date = df['Date'].max()
recent_data = df[df['Date'] >= (latest_date - pd.Timedelta(days=30))]

# Plotting
plt.figure(figsize=(14, 7))
for city in ['beijing', 'chengdu']:
    city_data = recent_data[recent_data['City'] == city].sort_values('Date')
    plt.plot(city_data['Date'], city_data['Water_Quality_Index'], marker='o', label=city.capitalize())

plt.title('Water Quality Index (Past Month)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Water Quality Index', fontsize=14)
plt.legend(title='City', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#From this time/WQI line graph, we can preliminarily analyze the changes of WQI of these two cities over time
#Beijing's WQI first drops, then rises, and then drops again, which seems to be cyclical, a bit like the image of sine and cosine functions
#Chengdu's WQI first rises, then falls, and then rises again, which also seems to be cyclical, a bit like the image of sine and cosine functions
```

```python
# From the line chart above, we can see that the WQI of the city seems to change in a cycle, so we select the seasonal decomposition chart to further observe its trend. Here we select Beijing as a sample.
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/lihaocong/Documents/Introtuction to Data Science/Beijing_water_pollution_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df = df.set_index('Date')
series = df['Water_Quality_Index'].asfreq('D').interpolate()

# Seasonal decomposition
decomposition = seasonal_decompose(series, model='additive', period=30)

# Plotting
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.suptitle("Seasonal Decomposition of WQI (Beijing)", fontsize=14)
plt.tight_layout()
plt.show()
```

## Regional Analysis

```python
# compare average WQI between different provence
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
df = pd.read_csv('/Users/lihaocong/Documents/Introtuction to Data Science/china_water_pollution_data.csv')

# data cleaning
df['Province'] = df['Province'].astype(str)
df_clean = df.dropna(subset=['Province', 'Water_Quality_Index'])

# calculate average WQI of different provence
province_avg = df_clean.groupby('Province')['Water_Quality_Index'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=province_avg.values, y=province_avg.index, palette=sns.color_palette("tab10", len(province_avg)))
plt.title('Average Water Quality Index by Province')
plt.xlabel('Average Water Quality Index')
plt.ylabel('Province')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

From this chart we can roughly see the WQI level of each province, and thus roughly see the difference in water quality among provinces in China
Overall, the water quality of these 10 provinces is not much different, all around 50
Generally speaking, WQI below 50 is excellent and 50-100 is good, indicating that the water quality of these 10 provinces is still good

```python
import pandas as pd

# Loading dataset
df = pd.read_csv("/Users/lihaocong/Documents/Introtuction to Data Science/Beijing_water_pollution_data.csv")

# Rename columns to match Prophet format
df_prophet = df.rename(columns={'Date': 'ds', 'Water_Quality_Index': 'y'})

# Drop NaNs if any
df_prophet = df_prophet.dropna(subset=['ds', 'y'])
```

```python
from prophet import Prophet
# Initialize and Train the Prophet Model
# Create a new Prophet model instance
model = Prophet()
model.fit(df_prophet)
```

```python
# Create future dates
future = model.make_future_dataframe(periods=30)  # extend 30 days ahead

# Make prediction
forecast = model.predict(future)
```

```python
from prophet.plot import plot_plotly

# Interactive plot
plot_plotly(model, forecast)
```

```python
model.plot_components(forecast)
```
